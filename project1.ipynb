```python
import csv
file_path = '/Users/yutonomy/Dropbox/Columbia SIPA/24FALL/Computing in Context/Project1/Pregnancy-Associated_Mortality_20241108.csv'
```

```python
with open(file_path, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        if row['Deaths']:  
            deaths_list.append(float(row['Deaths']))

mean_deaths_std = sum(deaths_list) / len(deaths_list) #sum of row /length of row = mean

sorted_deaths_list = sorted(deaths_list)

n = len(sorted_deaths_list)

if n % 2 == 0:
    median_deaths_std = (sorted_deaths_list[n // 2 - 1] + sorted_deaths_list[n // 2]) / 2 #if number in list is even, caliculate the mean of the 2 number on the center
else:
    median_deaths_std = sorted_deaths_list[n // 2] #if number in list is odd, pick the number on the center

frequency_dict = {}
for value in deaths_list:
    if value in frequency_dict:
        frequency_dict[value] += 1
    else:
        frequency_dict[value] = 1

max_count = max(frequency_dict.values()) #looking for 
mode_deaths_std = [key for key, count in frequency_dict.items() if count == max_count][0]

print("Mean:", mean_deaths_std)
print("Median:", median_deaths_std)
print("Mode:", mode_deaths_std)
```

```python
sorted_counts = sorted(frequency_dict.items()) #sorting results

print("Distribution of Deaths:") #data visualization
for value, count in sorted_counts:
    bar = '█' * (count // 2) 
    print(f"{value}: {bar}")
```
